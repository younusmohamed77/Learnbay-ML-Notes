```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('orders.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset['date'].describe()
```

```python
# convert object to datetime format
dataset['date'] = pd.to_datetime(dataset['date'])
```

```python
dataset.info()
```

```python
dataset['date'].describe()
```

```python
dataset.head()
```

```python
# part 1 - extract year and make new variable
dataset['years'] = dataset['date'].dt.year
```

```python
dataset.head()
```

```python
# part 2 - extract month and make it new variable
dataset['month'] = dataset['date'].dt.month
```

```python
dataset.head()
```

```python
dataset['month_name'] = dataset['date'].dt.month_name()
```

```python
dataset.head()
```

```python
# part 3 - extract day and make it new variable
dataset['day'] = dataset['date'].dt.day
```

```python
dataset.head()
```

```python
# day of the week

dataset['day_of_week'] = dataset['date'].dt.dayofweek
dataset.head()
```

```python
dataset['days_name'] = dataset['date'].dt.day_name()
dataset.head(10)
```

```python
# To find the weekend and weed day 

dataset['check_weekend'] = np.where(dataset['days_name'].isin(['Saturday','Sunday']),1,0)
dataset.head(10)
```

```python
dataset['check_weekend_char'] = np.where(dataset['days_name'].isin(['Saturday','Sunday']),'weekend','weekday')
dataset.head(10)
```

```python
dataset = dataset.drop(['product_id', 'city_id', 'orders'], axis=1)
dataset.head()
```

```python
dataset['week_of_year'] = dataset['date'].dt.week
dataset.head()
```

```python
dataset['Quarter_Report'] = dataset['date'].dt.quarter
dataset.head()
```

```python
dataset['half_yearly'] = np.where(dataset['Quarter_Report'].isin([1,2]),1,2)
dataset.head()
```
